```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from textblob import TextBlob
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Download NLTK stopwords
nltk.download("stopwords")
stop_words = set(stopwords.words("english"))
```

```python
data = {
    "Movie": ["Inception", "Titanic", "Avengers", "Joker", "Interstellar"],
    "Reviews": [
        "Mind-blowing and amazing movie! Great plot.",
        "A tragic yet beautiful love story. Heartbreaking.",
        "Full of action and excitement! A must-watch.",
        "Dark, thrilling, and well-acted movie. Joaquin Phoenix is brilliant.",
        "Visually stunning and deeply emotional. A masterpiece!"
    ]
}
```

```python
# Convert dataset to DataFrame
df = pd.DataFrame(data)

# Function to clean and preprocess reviews
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    words = text.split()  # Tokenize
    words = [word for word in words if word not in stop_words]  # Remove stopwords
    return " ".join(words)
```

```python
# Apply preprocessing
df["Processed_Reviews"] = df["Reviews"].apply(preprocess_text)

# Function to analyze sentiment of reviews
def analyze_sentiment(text):
    sentiment = TextBlob(text).sentiment.polarity  # Returns a score between -1 to 1
    return sentiment

# Apply sentiment analysis
df["Sentiment_Score"] = df["Processed_Reviews"].apply(analyze_sentiment)
```

```python
# Sort movies based on sentiment score (higher = more positive)
df_sorted = df.sort_values(by="Sentiment_Score", ascending=False)

# Movie recommendation function
def recommend_movies():
    print(" Recommended Movies Based on Positive Reviews:\n")
    for index, row in df_sorted.iterrows():
        print(f" {row['Movie']} (Sentiment Score: {round(row['Sentiment_Score'], 2)})")

# Display recommended movies
recommend_movies()
```
